```python
import pandas as pd
import numpy as np

#Reading data set
df = pd.read_csv(r'C:\\Users\\Purav\\Puravee_UCD Project\BostonHousing.csv')
df.head()
#penguins =pd.read_csv(r'C:\\Users\\Purav\\Puravee_UCD Project\penguins.csv')
#penguins.head()
```

```python
# Set up the dependent and the independent variables

x= pd.DataFrame(df.iloc[:,:-1])
y= pd.DataFrame(df.iloc[:,-1])
```

```python
# A glance at the independent variable
x
```

```python
# A glance at the independent variable
y
```

```python
#Divide the data into train and test sets:

from sklearn.model_selection import train_test_split
x_train , x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=5) 

```

```python
#Have a glance at the shape of the train and test sets:
print (x_train.shape)
print (x_test.shape)
print (y_train.shape)
print (y_test.shape)
 
```

```python
#Train the algorithm:

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#Having a look at the coefficients that the model has chosen:

v = pd.DataFrame(regressor.coef_ ,index= ['Co-efficient']).transpose()
w = pd.DataFrame(x.columns, columns = ['attributes'])
```

```python
#Concatenating the DataFrames to compare:
coeff_df = pd.concat([w,v], axis=1, join='inner')
coeff_df
```

```python
#Comparing the predicted value to the actual value:

y_pred = regressor.predict(x_test)
y_pred = pd.DataFrame(y_pred, columns=['Predicted'])
y_pred

```

```python
y_test
```

```python
#Evaluate the algorithm

from sklearn import metrics
print ('Mean Absolute Error:' , metrics.mean_absolute_error(y_test , y_pred))
print ('Mean Squared Error:' , metrics.mean_squared_error(y_test , y_pred))
print ('Root Mean Squared Error:' , np.sqrt(metrics.mean_squared_error(y_test , y_pred)))
```
